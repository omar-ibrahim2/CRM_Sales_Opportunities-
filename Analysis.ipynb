```python
-- How is each sales team performing compared to the rest?
select count(case when sp.deal_stage = 'won' then 1 else 0 end) as total_wins ,
    st.sales_agent,
    st.regional_office,
    SUM(sp.close_value) as Sales
   from "sales_pipeline.csv" sp
join "sales_teams.csv" st
   on sp.sales_agent = st.sales_agent
group by st.sales_agent, sp.deal_stage ,st.regional_office 
order by Sales desc
```

```python
--Are there any quarter-over-quarter trends?
select engage_date 
	,close_date 
	,quarter(close_date) as Quarter
    , sum(close_value) as Sales  
from 'sales_pipeline.csv'
group by engage_date,Quarter,close_date 
order by  Sales desc
```

```python
--Do any products have better win rates?
select year(close_date) as year
	,product
	,deal_stage
	,sum(close_value) as Sales
     from  'sales_pipeline.csv' 
group by product,deal_stage,year(close_date)  
order by Sales desc
```

```python
--Avg sales from every sector 
select acc.sector
      ,avg(sp.close_value) as sales
      from 'accounts.csv' acc
join 'sales_pipeline.csv' sp 
	on acc.account = sp.account
group by sector
order by sales desc 
```

```python
--Relationship between  size of the company with revenue
select account 
      ,employees
	  , revenue
      from 'accounts.csv'
order by revenue desc
```

```python
--Relationship between the region offices with their sales and the number of employees
select st.regional_office
	,sum(sp.close_value) as Sales
    ,count(st.sales_agent) as employee
    from "sales_pipeline.csv" sp
join "sales_teams.csv" st
   on sp.sales_agent = st.sales_agent
group by st.regional_office
order by Sales desc 
```
